Nhận xét: Để xử lý nhiễu, ta chọn phương pháp thay thế nhiễu bằng tham số median. Quan sát thấy các outlier đều nằm ở bên phải Q3 (Q3 bằng với phân vị của 75%) nên ta sẽ tính chênh lệch tứ phân vị và tính giới hạn trên vừa phải (Reasonable Upper Bound). Nếu giá trị của thuộc tính kiểm tra lớn hơn giá trị giới hạn trên thì thay thế giá trị vừa kiểm tra bằng **median**

```python
##Tính giới hạn trên
column = numerical_data_shoppers.columns
q1_list = []
q3_list = []
medianlist = []
for i in column:
  i1 = data_shoppers[i].quantile(.25)
  i3 = data_shoppers[i].quantile(.75)
  i2 = data_shoppers[i].quantile(.5)
  q1_list.append(i1)
  q3_list.append(i3)
  medianlist.append(i2)
q3_list[4] = 38
upper_bound = [(x+(x-y)*1.5) for x,y in zip(q3_list,q1_list)]
```

```python
##Kiểm tra và thay thế giá trị
data_shoppers['Administrative'] = np.where(data_shoppers['Administrative']>upper_bound[0],medianlist[0],data_shoppers['Administrative'])
data_shoppers['Administrative_Duration'] = np.where(data_shoppers['Administrative_Duration']>upper_bound[1],medianlist[1],data_shoppers['Administrative_Duration'])
data_shoppers['Informational'] = np.where(data_shoppers['Informational']>upper_bound[2],medianlist[2],data_shoppers['Informational'])
data_shoppers['Informational_Duration'] = np.where(data_shoppers['Informational_Duration']>upper_bound[3],medianlist[3],data_shoppers['Informational_Duration'])
data_shoppers['ProductRelated'] = np.where(data_shoppers['ProductRelated']>upper_bound[4],medianlist[4],data_shoppers['ProductRelated'])
data_shoppers['ProductRelated_Duration'] = np.where(data_shoppers['ProductRelated_Duration']>upper_bound[5],medianlist[5],data_shoppers['ProductRelated_Duration'])
data_shoppers['BounceRates'] = np.where(data_shoppers['BounceRates']>upper_bound[6],medianlist[6],data_shoppers['BounceRates'])
data_shoppers['ExitRates'] = np.where(data_shoppers['ExitRates']>upper_bound[7],medianlist[7],data_shoppers['ExitRates'])
data_shoppers['PageValues'] = np.where(data_shoppers['PageValues']>upper_bound[8],medianlist[8],data_shoppers['PageValues'])
data_shoppers['SpecialDay'] = np.where(data_shoppers['SpecialDay']>upper_bound[9],medianlist[9],data_shoppers['SpecialDay'])
```

```python
data = data_shoppers_processed
```

```python
to_drop = ["Administrative", "Informational", "ProductRelated", "PageValues", "SpecialDay", "Month","OperatingSystems",
           "Browser","Region","TrafficType","VisitorType","Weekend","Revenue"]
data_3 = data.drop(to_drop, axis=1)
data_3.head()
```

####2.1.1.Tìm số cụm K theo phương pháp Elbow

```python
print('Phương pháp Elbow để xác định số lượng cụm được hình thành:')
Elbow_M = KElbowVisualizer(KMeans(), k=(1,10))
Elbow_M.fit(data_3)
Elbow_M.show()
```

```python
#Chọn k dựa trên Elbow Method
k = 3
model_kMeans = KMeans(n_clusters = k)
y_means = model_kMeans.fit(data_3)

## Các clusters
labels     = model_kMeans.labels_
clustering = pd.concat([data_3, pd.Series(labels, name = 'cluster')], axis = 1)
print('Số phần tử của mỗi cluster:')
print(clustering.cluster.value_counts(), '\n')

## Các trọng tâm = các vectors trong không gian 4 chiều
print(f'Tọa độ của {k} trọng tâm:')
centroids  = model_kMeans.cluster_centers_
print(centroids)
```

####2.1.2.Biểu diễn kết quả gom cụm

```python
plt.figure(figsize = (20, 20))
plt.subplot(2, 2, 1)
sns.scatterplot(data= clustering, x = 'ProductRelated_Duration', y = 'BounceRates', hue = 'cluster')
sns.scatterplot(data= centroids,  x = centroids[:, 2], y = centroids[:, 3], marker = 's', color = 'r')
plt.show()
```

 ExitRates vs Administratice_Duration plot cho ra các cụm:
- group 0: tỷ lệ thoát mức trug bình & thời gian dành cho việc truy cập mức trung bình (nhóm khách hàng chung)
- group 1: tỷ lệ thoát cao & thời gian dành cho việc truy cập thấp (nhóm khách hàng không hứng thú với mua hàng - khả năng giao dịch thấp)
- group 2: tỷ lệ thoát thấp & thời gian dành cho việc truy cập cao (nhóm khách hàng mục tiêu - khả năng giao dịch cao)

####2.1.3.Tính điểm Silhoutte

```python
# Tính toán Silhoutte Score để đánh giá phân cụm có đang đúng hay không
score = silhouette_score(data_3, model_kMeans.labels_, metric='euclidean')
# Silhouette score từ 0.5 trở lên, bằng chứng cho thấy có thể cluster này sát với thực tế
print('Silhouetter Score: %.3f' % score)
```

###3.XÂY DỰNG MÔ HÌNH XÁC ĐỊNH TẬP CÁC LỚP DỮ LIỆU

#####3.1. Tập huấn luyện

```python
# Tách biến target
y = data_shoppers_ml.Revenue
x = data_shoppers_ml.drop(['Revenue'], axis = 1)
```

```python
# Checking the shape
print('Shape x',x.shape)
print('Shape y',y.shape)
```

```python
# Stratified sampling
# Bắt đầu lấy mẫu phân tầng
x_train_stratified, x_test_stratified, y_train_stratified, y_test_stratified = train_test_split(x, y,stratify=y, test_size=.2, random_state=2, shuffle=True)
```

```python
# Checking the shape
print('Shape x_train_stratified',x_train_stratified.shape)
print('Shape y_train_stratified',y_train_stratified.shape)
print('Shape x_test_stratified',x_test_stratified.shape)
print('Shape y_test_stratified',y_test_stratified.shape)
```

#####3.2. Xây dựng mô hình

######a. Logistic Regression

```python
classifier = LogisticRegression()

lr = classifier.fit(x_train_stratified, y_train_stratified)

y_pred_lr_stratified = lr.predict(x_test_stratified)
```

Confusion matrics

```python
from sklearn.metrics import confusion_matrix
```

```python
conmat = confusion_matrix(y_test_stratified, y_pred_lr_stratified)
val = np.mat(conmat)

classnames = list(set(y_train_stratified))

df_cm = pd.DataFrame(val, index=classnames, columns=classnames)

print(df_cm)

df_cm = df_cm.astype('float') / df_cm.sum(axis=1)[:, np.newaxis]

plt.figure()

heatmap = sns.heatmap(df_cm, annot=True, cmap="Blues")

heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right')

heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right')

plt.ylabel('True label')

plt.xlabel('Predicted label')

plt.title('Churn Stratified Logistic Regression Model Results')

plt.show()


report = classification_report(y_test_stratified, y_pred_lr_stratified)
print('report', report, sep = '\n')
```

*ĐÁNH GIÁ*

Sử dụng thuật toán Logistic Regression để xây dựng mô hình dự đoán cho chúng ta độ chính xác khoảng 84%. Chỉ số Precision của trường hợp False lên tới 85% trong khi True chỉ là 26%. Tiếp tục ta sẽ thực hiện với nhiều thuật toán khác để có thể tìm ra được thuật toán phù hợp nhất.
